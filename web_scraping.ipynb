```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "........"
req = requests.get(url)
soup = BeautifulSoup(req.text, "html.parser")
liste=list()
for link in soup.find_all("a",attrs={"class":"html5lightbox"}):
    link=link.get('href')
    liste.append([link])
del liste[0:2]
print(liste)
```
